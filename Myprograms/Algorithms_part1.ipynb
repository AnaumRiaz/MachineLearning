```python
#SHELL SORT

def shellsort(gap, values):
    siz = len(values)
    for i in range(gap, siz):
        checker = i
        while (checker) >= gap:
            if values[checker] < values[checker-gap]:

                interim = values[checker - gap]
                values[checker - gap] = values[checker]
                values[checker] = interim
                print(values)
                checker -= gap
            else:
                break
    return values
    

array = [61, 55, 48, 39, 26, 14, 77, 0, 1, 3, 23, 9,25,0,23]
print(array)
array2 = shellsort(3,array)
print(array2)

```

```python
#mergesort

import math

def merge_sort(values):
    sz = len(values)
    #basecase
    if sz == 1:
        return values

    #split the array in two if its not basecase
    mid = math.ceil(sz/2)
    left_half = values[:mid]
    right_half = values[mid:]
    
    #sort them 
    #here, recursively, calling merge_sort on left half will keep going until basecase.
    #after basecase, it goes one step up, do the right_half and then merge. then it goes one step up
    #and does the same thing. so it goes through the left half first and keeps going until it gets the answer,
    #then it moves on to right half and merge
    print('left_half', left_half, 'right_half', right_half)

    sorted_l = merge_sort(left_half)
    sorted_r = merge_sort(right_half)
    
    
    mergesorted = merge(sorted_l, sorted_r)
    return mergesorted

def merge(left_half, right_half):
    print('merging', left_half, right_half)
    interim_list = []
    left_pointer = 0
    right_pointer = 0
    sz_l = len(left_half)
    
    
    #when merging, it checks which array element is bigger. then moves that pointer and keeps going
    #Stop if already sorted.・Is biggest item in first half ≤ smallest item in second half
    if left_half[sz_l-1] <= right_half[0]:
        interim_list.extend(left_half)
        interim_list.extend(right_half)
    
    else:    
        while left_pointer < len(left_half) and right_pointer < len(right_half):
            if left_half[left_pointer] < right_half[right_pointer]:
                interim_list.append(left_half[left_pointer])
                left_pointer+=1
        
            else:
                interim_list.append(right_half[right_pointer])
                right_pointer+=1
        #using extend here will copy the array in the new list like [1,2,3,[8]] instead of [1,2,3,8]  
        #once, one of the pointer is exhausted, meaning while sorting, we have shifted one
        #half(left or right) to the sorted interim array, so no more comparisons needed. we just copy the rest of the 
        #array elements
        interim_list.extend(left_half[left_pointer:])
        interim_list.extend(right_half[right_pointer:])
    
    return interim_list
        
    

array = [61, 55, 48, 39, 26, 14, 77, 0, 1, 3, 23, 9,25,0,23]
print(array)
sorted_ = mergesort(array)
sorted_
```

```python
#quicksort
import random    #for shuffling


def partition(array, start, end):
    print('inside partition with start',start, 'and end', end)
    i, j = start+1, end
    
    while (start<end):
        
        #checking left pointer until we encounter an element that is bigger
        while(array[i] < array[start]):
            print('inside first loop with i',i)
            if i>=end:
                print('i ran outttt')
                break
            
            i+=1
       
        
        #Now we start the j pointer until we find element that is smaller than array[pkey]
        while(array[j] > array[start]):
            print('inside second loop with j', j)
            if (j <= start):
                print('j is less than start.')
                break
            j-=1
        
        #exchange only if i is less than j. if they are equal, makes no sense to exchange and if i >j, 
        #again pointers have crossed
        if i<j:
            print('before exchnage, array is ', array)
            tmp = array[i]
            array[i] = array[j]
            array[j]= tmp
            print('after exchnage, array is ', array)
            i+=1
            j-=1
        
        #the above will go on until the pointers cross each other or i/j loops are exhausted
        else:# i>=j:
            print('POINTERS CROSSING')
            break
            
    print('bringing the partition element', array[start],'to index', j)
    tmp = array[j]
    array[j] = array[start]
    array[start]= tmp
    print('final array', array)
    return j
        
#now we will use partition to sort
def quicksort(array, start ,end):
    
    if start >= end:
        return
    
    pindex = partition(array, start, end)
    
    quicksort(array, 0, pindex-1)
    quicksort(array, pindex + 1, end)
    
    
r_array = [5,8,85,4,5,6,99,1,3,12,9,0,4,1,8,2,15]   

random.shuffle(r_array)
print('original shuffled', r_array) 
quicksort(r_array, 0, len(r_array) -1)
```

```python
#nuts and bolts problem
   
def exch(array, ind1, ind2):
    tmp = array[ind1]
    array[ind1] = array[ind2]
    array[ind2] = tmp
    return array
    
def quicksort_3way(array, start, end):
    if start >= end:
        return
    
    #taking care of three pointers, lt always being the partition element
    lt, gt = start, end
    i = start+1
    
    while (i<=gt):
        
        if array[i]<array[lt]:
            array = exch(array, i, lt)
            i+=1
            lt+=1
            
        elif array[i] > array[lt]:
            array = exch(array, i, gt)
            gt-=1
            
        else: 
            i+=1
    
    #deleting one of the element because we know two elements are same(nut and bolt)
    del array[lt]
    
    return array, lt
      
nuts = [5,10,9,8,7,4,3,1,2,12,14,6]
bolts =[1,5,10,8,7,9,6,14,4,2,12,3]

#Here end is index + 1 because we insert one item before quicksort
def nuts_and_bolts(nuts, bolts, start, end):
    #end case for recursion.
    if start >= end-1 or end < 0:
        return

    #pick first element of nuts and sort bolts with it
    bolts.insert(start, nuts[start])
    bolts, index = quicksort_3way(bolts, start, end)
    print('bolts:',bolts)

    #with the returned index(we know that bolts[index] = nuts[0]. index returned the right sorted place in the aray)
    nuts.insert(start, bolts[index])
    nuts, index = quicksort_3way(nuts, start, end)
    print('Nuts:',nuts)
   
    nuts_and_bolts(nuts,bolts, start, index)
    nuts_and_bolts(nuts, bolts, index+1, end)


start = 0 
#here this end works because of end -1 because we append another value inside the loop
end = len(nuts)
nuts_and_bolts(nuts,bolts, start, end)

print('nuts', nuts)
print('bolts', bolts)
```

```python
class HashTable(object):
    def __init__(self, length=4):
        # Initiate our array with empty values.
        self.array = [None] * length
    
    def hash(self, key):
        """Get the index of our array for a specific string key"""
        length = len(self.array)
        return hash(key) % length
        
    def add(self, key, value):
        """Add a value to our array by its key"""
        if self.is_full():
            self.double()
            
        print('Adding key value', key, value)
        index = self.hash(key)
        
        if self.array[index] is not None:
            # This index already contain some values.
            # This means that this add MIGHT be an update
            # to a key that already exist. Instead of just storing
            # the value we have to first look if the key exist.
            for kvp in self.array[index]:
                # If key is found, then update
                # its current value to the new value.
                if kvp[0] == key:
                    kvp[1] = value
                    break
            else:
                # If its not in the list in our index, add.
                self.array[index].append([key, value])
        else:
            # This index is empty. We should initiate 
            # a list and append our key-value-pair to it.
            self.array[index] = []
            self.array[index].append([key, value])
    
    def get(self, key):
        """Get a value by key"""
        index = self.hash(key)
        if self.array[index] is None:
            raise KeyError()
        else:
            # Loop through all key-value-pairs
            # and find if our key exist. If it does 
            # then return its value.
            for kvp in self.array[index]:
                if kvp[0] == key:
                    return kvp[1]
            
            # If no return was done during loop,
            # it means key didn't exist.
            raise KeyError()
            
    def all_vals(self):         
        items = 0
        for item in self.array:
            if item is not None:
                print(item)
                print('------')
            else:
                print ('Empty')
            
    def is_full(self):
        """Determines if the HashTable is too populated. here it means its more than half full"""
        items = 0
        # Count how many indexes in our array are full
        for item in self.array:
            if item is not None:
                items += 1
        # Return true if more than half full
        return items > len(self.array)/2
        
    def double(self):
        """Double the list length and re-add values"""
        print('doubling the array')
        ht2 = HashTable(length=len(self.array)*2)
        for i in range(len(self.array)):
            if self.array[i] is None:
                continue
            
            # Since our list is now a different length,
            # we need to re-add all of our values to 
            # the new list for its hash to return correct
            # index.
            for kvp in self.array[i]:
                ht2.add(kvp[0], kvp[1])
        
        # Finally we just replace our current list wih the new list of values that we created in ht2.
        self.array = ht2.array

mh = HashTable()
mh.add(5, 'five')
mh.add(6, 'six')
mh.add(7, 'seven')
mh.add(8, 'eight')
mh.add(9, 'nine')
mh.add(10, 'ten')
mh.add(11, 'eleven')
```
