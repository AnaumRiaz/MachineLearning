```python
#SHELL SORT

def shellsort(gap, values):
    siz = len(values)
    for i in range(gap, siz):
        checker = i
        while (checker) >= gap:
            if values[checker] < values[checker-gap]:

                interim = values[checker - gap]
                values[checker - gap] = values[checker]
                values[checker] = interim
                print(values)
                checker -= gap
            else:
                break
    return values
    

array = [61, 55, 48, 39, 26, 14, 77, 0, 1, 3, 23, 9,25,0,23]
print(array)
array2 = shellsort(3,array)
print(array2)

```

```python
#mergesort

import math

def merge_sort(values):
    sz = len(values)
    #basecase
    if sz == 1:
        return values

    #split the array in two if its not basecase
    mid = math.ceil(sz/2)
    left_half = values[:mid]
    right_half = values[mid:]
    
    #sort them 
    #here, recursively, calling merge_sort on left half will keep going until basecase.
    #after basecase, it goes one step up, do the right_half and then merge. then it goes one step up
    #and does the same thing. so it goes through the left half first and keeps going until it gets the answer,
    #then it moves on to right half and merge
    print('left_half', left_half, 'right_half', right_half)

    sorted_l = merge_sort(left_half)
    sorted_r = merge_sort(right_half)
    
    
    mergesorted = merge(sorted_l, sorted_r)
    return mergesorted

def merge(left_half, right_half):
    print('merging', left_half, right_half)
    interim_list = []
    left_pointer = 0
    right_pointer = 0
    sz_l = len(left_half)
    
    
    #when merging, it checks which array element is bigger. then moves that pointer and keeps going
    #Stop if already sorted.・Is biggest item in first half ≤ smallest item in second half
    if left_half[sz_l-1] <= right_half[0]:
        interim_list.extend(left_half)
        interim_list.extend(right_half)
    
    else:    
        while left_pointer < len(left_half) and right_pointer < len(right_half):
            if left_half[left_pointer] < right_half[right_pointer]:
                interim_list.append(left_half[left_pointer])
                left_pointer+=1
        
            else:
                interim_list.append(right_half[right_pointer])
                right_pointer+=1
        #using extend here will copy the array in the new list like [1,2,3,[8]] instead of [1,2,3,8]  
        #once, one of the pointer is exhausted, meaning while sorting, we have shifted one
        #half(left or right) to the sorted interim array, so no more comparisons needed. we just copy the rest of the 
        #array elements
        interim_list.extend(left_half[left_pointer:])
        interim_list.extend(right_half[right_pointer:])
    
    return interim_list
        
    

array = [61, 55, 48, 39, 26, 14, 77, 0, 1, 3, 23, 9,25,0,23]
print(array)
sorted_ = mergesort(array)
sorted_
```
