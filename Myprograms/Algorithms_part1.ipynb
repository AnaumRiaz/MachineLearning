```python
#implementing queue
#implementing stack using linkedlist
#implement push and pop

#Linkedlist
class Node:
    
    '''This is a docstring. I have created a new class modeling nodes'''
    def __init__(self, dataval=None):
        self.data = dataval
        self.nextnode = None
        
    def __repr__(self):
        return str(self.data)
    
class QLinkedList:
    
    '''This is a docstring. This class makes linkedlist'''
    def __init__(self, values=None):
        if values is not None:
            self.size = len(values)
            self.head = Node(values.pop(0))
            current = self.head
            for value in values:
                current.nextnode = Node(value)
                current = current.nextnode  
        else:
            self.head = None
            self.size = 1
        
    def __repr__(self):
        allnodes = []
        current = self.head
        
        while current is not None:
            allnodes.append(str(current.data))
            current = current.nextnode
        allnodes.append('None')
        return " -> ".join(allnodes)
    
    #without this, it will say the thing is not iterable(when using for loops.)
    def __iter__(self):
        current = self.head
        while current is not None:
            yield current
            current = current.nextnode
            
    #pushing the value to the start of the linkedlist    
    def push(self, value=None):
            current = self.head
            self.head = Node(value)
            self.head.nextnode = current
            self.size += 1
            
    def pop(self):
        current = self.head
        while current.nextnode is not None:
            current = current.nextnode
        self.delete_node(self.size - 1)
        return current
        
    def delete_node(self, index):
        if index == 0:
            self.head = self.head.nextnode
            self.size -= 1
            
        elif index > self.size+1:
            print('index greater than the length of linkedlist')
        
        else:
            prev = index - 1
            current = self.head
            
            for i in range(index):
                if i == prev:
                    prev_node = current
                current = current.nextnode
            
            prev_node.nextnode = current.nextnode  
            self.size -= 1
```
